```python
from pandas.api.types import CategoricalDtype
```

```python
departments = pd.Series(["HR", "Finance", "IT", "Marketing"], dtype="category")
```

```python
gender = pd.Series([1, 0, 1, 1]) 
pass_fail = pd.Series(["Pass", "Fail", "Pass", "Pass"], dtype="category")
```

```python
priority_type = CategoricalDtype(categories=["Low", "Medium", "High"], ordered=True)
priority = pd.Series(["Low", "High", "Medium", "Low"], dtype=priority_type)
```

```python
num_of_projects = pd.Series([3, 5, 2, 4], dtype="int")
```

```python
salary = pd.Series([45000.50, 54000.00, 61000.75, 72000.10], dtype="float")
```

```python
df = pd.DataFrame({
    "Department (Nominal)": departments,
    "Gender (Binary)": gender,
    "Pass/Fail (Binary)": pass_fail,
    "Priority (Ordinal)": priority,
    "Projects (Discrete Numeric)": num_of_projects,
    "Salary (Continuous Numeric)": salary
})
```

```python
print(df)
print("\nData Types:\n", df.dtypes)
```

```python
import pandas as pd
data = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Edward'],
    'Department': ['HR', 'IT', 'IT', 'Marketing', 'Sales'],
    'Age': [29, 34, 41, 28, 38],
    'Salary': [50000, 70000, 65000, 55000, 60000],
    'Years_of_Experience': [4, 8, 10, 3, 12],
    'Joining_Date': ['2020-03-15', '2017-07-19', '2013-06-01', '2021-02-10', '2010-11-25'],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Male'],
    'Bonus': [5000, 7000, 6000, 4500, 5000],
    'Rating': [4.5, 4.0, 3.8, 4.7, 3.5]
}
```

```python
df = pd.DataFrame(data)

# a) Shape of DataFrame
print("Shape:", df.shape)
```

```python
# b) Summary: Data types and non-null counts
print("\nDataFrame Info:")
print(df.info())
```

```python
# c) Descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
# d) First 5 rows and last 3 rows
print("\nFirst 5 rows:")
print(df.head())
print("\nLast 3 rows:")
print(df.tail(3))
```

```python
# e) Calculations & statistics
print("\nAverage Salary:", df['Salary'].mean())
print("Total Bonus:", df['Bonus'].sum())
print("Youngest Age:", df['Age'].min())
print("Highest Performance Rating:", df['Rating'].max())
```

```python
# f) Sort by salary descending
df_sorted = df.sort_values(by='Salary', ascending=False)
print("\nSorted by Salary (descending):")
print(df_sorted)
```

```python
# g) New column: Performance Category
def categorize_rating(r):
    if r >= 4.5:
        return "Excellent"
    elif r >= 4.0:
        return "Good"
    else:
        return "Average"
df['Performance_Category'] = df['Rating'].apply(categorize_rating)
print("\nWith Performance Category:")
print(df[['Name', 'Rating', 'Performance_Category']])
```

```python
# h) Identify missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# i) Rename Employee_ID to ID
df.rename(columns={'Employee_ID': 'ID'}, inplace=True)
```

```python
# j) Filter employees
over_5_years = df[df['Years_of_Experience'] > 5]
print("\nEmployees with >5 years experience:")
print(over_5_years)
it_department = df[df['Department'] == 'IT']
print("\nEmployees in IT Department:")
print(it_department)
```

```python
# k) Add Tax column: 10% of Salary
df['Tax'] = df['Salary']*0.1
```

```python
# l) Save to new CSV file
df.to_csv('employees_modified.csv', index=False)
```
