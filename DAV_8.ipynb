```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import datasets

d = datasets.load_wine()
X = d.data
y = d.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=6)

clf = DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
from collections import defaultdict, Counter
import numpy as np

def oneR(X, y):
    n_samples, n_features = X.shape
    min_error = float('inf')
    best_feature = None
    best_rules = None

    for feature in range(n_features):
        rules = {}
        error = 0

        values = np.unique(X[:, feature])

        for val in values:
            indices = np.where(X[:, feature] == val)[0]
            class_counts = Counter(y[indices])
            if class_counts:
                predicted_class, count = class_counts.most_common(1)[0]
                error += len(indices) - count
                rules[val] = predicted_class
                
        if error < min_error:
            min_error = error
            best_feature = feature
            best_rules = rules

    return best_feature, best_rules

def predict_oneR(X, feature, rules, default_class):
    y_pred = []
    for x in X[:, feature]:
        y_pred.append(rules.get(x, default_class))
    return np.array(y_pred)

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

w = datasets.load_breast_cancer()
X = w.data
y = w.target

X_discrete = np.array(X)
for i in range(X.shape[1]):
    X_discrete[:, i] = np.digitize(X[:, i], bins=np.histogram_bin_edges(X[:, i], bins=5))

X_train, X_test, y_train, y_test = train_test_split(X_discrete, y, test_size=0.2)

feature, rules = oneR(X_train, y_train)
default_class = Counter(y_train).most_common(1)[0][0]

y_pred = predict_oneR(X_test, feature, rules, default_class)
print(f"Best feature index: {w.feature_names[feature]}")
print(f"Rules: {rules}")
print("Accuracy:", accuracy_score(y_test, y_pred))
```
