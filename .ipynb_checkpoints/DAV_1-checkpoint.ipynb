```python
n=int(input("Enter a number: "))
s=0
for i in range(1,n):
	for j in range(2,i):
		if i%j==0:
			break
	s+=i
print(s)
```

```python
def prime(n):
    for i in range(2,n):
        if n%i==0:
            return 0
    return 1
def total(num): 
    s=0
    for j in range (1,num+1):
        if prime(j):
            s=s+j
    return s
n=int(input("Enter a number: "))
print(total(n))
```
